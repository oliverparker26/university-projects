```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
import datetime
from datetime import datetime
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
table = pd.DataFrame()
stock = 'EBAY'
money = 100000
stop_loss = 7
start = datetime(2015, 1, 1)
end = datetime(2019, 12,31)

df = web.get_data_yahoo(stock, start=start, end=end)
df = df[['Open', 'High', 'Low', 'Close']]
short_window = df['Close'].rolling(30).mean()
long_window = df['Close'].rolling(90).mean()

dates = []
signals = []
prices = []
left = 0
number = 0
capitals = []
rests = []
numbers = []
share_price = []
cash = []
stop_loss_price = 0
stop_l = []
q = []

for i in range(0, len(short_window)):
    
            stop_l.append(stop_loss_price)
            if ((short_window[i-1] < long_window[i-1]) & (short_window[i] > long_window[i]) & (number == 0) & (df['Low'][i] > stop_loss_price)):
                rests.append(left)
                
                cash.append(np.round(money,2))
                
                numbers.append(number)
                                
                
                share_price.append(np.round(df['Close'][i],2))
                share_price.append(np.round(df['Open'][i+1],2))
                number = number + int(money / df['Open'][i+1])
                stop_loss_price = (1-stop_loss/100) * df['Open'][i+1]
                stop_l[i] = stop_loss_price
                prices.append(np.nan)
                prices.append(np.round(df['Open'][i+1] * number,2))
                dates.append(df.index[i])
                dates.append(df.index[i+1])
                signals.append('sig_buy')
                signals.append('buy')
                
                if (money - number*df['Open'][i+1] >= 0):
                    cash.append(np.round((left + money), 2))
                    left = np.round(left + (money - number*df['Open'][i+1]),2)
                else:
                    rests.append(0)
                    left = 0
                money = 0
                rests.append(np.round(left,2))
                numbers.append(number)
                
            if ((number != 0) & (df['Low'][i] <= stop_loss_price)):
                
                dates.append(df.index[i])
                share_price.append(np.round(stop_loss_price,2))
                prices.append(np.round(stop_loss_price * number + left,2))
                signals.append('stop_loss')
                money = stop_loss_price * number + left
                left = 0
                number = 0
                rests.append(np.round(left,2))
                cash.append(np.round(money,2))
                numbers.append(number)
                stop_loss_price = 0
            if ((short_window[i-1] > long_window[i-1]) & (short_window[i] < long_window[i]) & (number !=0)):
                rests.append(np.round(left,2))
                cash.append(np.round(left,2))
                numbers.append(number)
                share_price.append(np.round(df['Close'][i],2))
                share_price.append(np.round(df['Open'][i+1],2))
                prices.append(np.nan)
                prices.append(np.round(df['Open'][i+1] * number,2))
                dates.append(df.index[i])
                dates.append(df.index[i+1])
                signals.append('sig_sale')
                signals.append('sale')
                money = df['Open'][i+1] * number + left
                cash.append(np.round(df['Open'][i+1] * number + left,2))
                rests.append(np.round(left,2))
                left = 0
                number = 0
                numbers.append(number)
            if (number == 0):
                q.append(0)
            else: 
                q.append(1)
                
table['date'] = dates
table['signal'] = signals
table['num_shares'] = numbers
table['share_price'] = share_price
table["share_value"] = prices
table['rest'] = rests
table['cash'] = cash
stop_l.append(stop_loss_price)

if (signals[-1] == 'buy'):
        table = table.iloc[:-2]
        print(np.round('Your profit is', table.cash.values[-1] - table.cash.values[0]),2)
        print(np.round('In percentage:',
              (100 * (table.cash.values[-1] - table.cash.values[0]) - table.cash.values[0]) / table.cash.values[0],'%'),2)      
else:
        print('Your profit is', np.round(table.cash.values[-1] - table.cash.values[0], 2))
        print('In percentage:', np.round((100 * (table.cash.values[-1] - table.cash.values[0]) - table.cash.values[0]) / table.cash.values[0], 2), '%')

display(table)
```

```python
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.2)
                  
fig.add_trace(go.Candlestick(x=df.index,
                open=df.Open,
                high=df.High,
                low=df.Low,
                close=df.Close, name = stock), row=1, col=1)

fig.update_layout(xaxis_rangeslider_visible=True)
                  
fig.add_trace(go.Scatter(x=df.index, y=short_window,
                    mode='lines', name='short', line=dict(color='purple', width=2)))
fig.add_trace(go.Scatter(x=df.index, y=long_window,
                    mode='lines', name='long', line=dict(color='royalblue', width=2)))
fig.add_trace(go.Scatter(x=table[table['signal'] == 'sig_buy'].date, y = table[table['signal'] == 'sig_buy'].share_price,
                    mode='markers', name='sig_buy', marker_size=13, marker_symbol='triangle-up',
                         marker = dict(color='green')))
fig.add_trace(go.Scatter(x=table[table['signal'] == 'sig_sale'].date, y = table[table['signal'] == 'sig_sale'].share_price,
                    mode='markers', name='sig_sale', marker_size=13, marker_symbol='triangle-down',
                         marker = dict(color='red')))
fig.add_trace(go.Scatter(x=table[table['signal'] == 'stop_loss'].date, y = table[table['signal'] == 'stop_loss'].share_price,
                    mode='markers', name='stop loss', marker_size=13, marker_symbol='triangle-down',
                         marker = dict(color='black')))
fig.add_trace(go.Scatter(x=df.index, y=stop_l,
                    mode='lines', name='линия цены стоп лосса', line=dict(color='black', width=0.7)))

fig.add_trace(go.Scatter(x = df.index, y = q, line=dict(color='blue', width=2),
                         name='Do we have stocks'),
              row=2, col=1)

fig.update_layout(width=900,
        height=1000,
        title=stock,
        yaxis_title='Price'
        
        )

fig.show()
```
