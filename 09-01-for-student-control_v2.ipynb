```python
# 1 балл (сумма 8) 
# создать отдельную таблицу для зимних олимпийских игр с 1956 с 1988 в которой
# перрекодировать медали 'Gold':3 , 'Silver': 2, 'Bronze': 1
#
# 1 балл (сумма 9)
# найти СПИСОК 6 команд набравщих суммарно больше всех баллов в медальном зачете
#
# 1 балл (сумма 10)
# Найти в каком году и в каком городе было получено больше всего очков странами из данного списка. 
# Результат представить в виде таблицы (номер по порядку, страна из списка 6 стран, год, город, количество баллов). 
# Упорядочить по количеству набранных баллов.


dictt = {'Gold' : 3, 'Silver': 2, 'Bronze': 1}
new = df[(df.Year.isin(range(1956,1989)) & (df.Season == 'Winter'))]
new = new.replace({'Medal': {'Gold' : 3, 'Silver': 2, 'Bronze': 1}})
six = new.groupby('Team')['Medal'].sum().nlargest(6)
#new.Medal.unique()
#new.Medal.unique()
six.index
```

```python
def find(country):
    year = new[new.Team == country].groupby('Year').Medal.sum().nlargest(1).index[0]
    return (new[(new.Team == country) & (new.Year == year)].City.values[0])
find('Soviet Union')
def year(country):
    return (new[new.Team == country].groupby('Year').Medal.sum().nlargest(1).index[0])
```

```python
big = pd.DataFrame()
#big['Countries'] = six.index
big['Medals'] = six
big['Year'] = big.index.map(lambda x: year(x))
big['Cities'] = big.index.map(lambda x: find(x))


big = big.reset_index()
```
