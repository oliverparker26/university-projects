```python
import numpy as np
import pandas as pd
import re
import tabula
from tabula.io import read_pdf
pd.set_option('display.max_rows', None)
```

```python
df_pdf = read_pdf('kcp-academ.pdf', 
    pages= 'all',
    lattice=True,
    pandas_options={"header": [0, 1, 2, 3]},
    relative_area=True
                        )

df = df_pdf[0]
for df_temp in df_pdf[1:]:
    df = df.append(df_temp,ignore_index=True)

df_pdff = read_pdf('PP-2022-pril-2_fil.pdf', 
    pages= 'all',
    lattice=True,
    pandas_options={"header": [0, 1, 2, 3]},
    relative_area=True)

dff = df_pdff[0]
for df_temp in df_pdff[1:]:
    dff = dff.append(df_temp,ignore_index=True)
```

```python
first = df[df[0].str.contains('СПЕЦИАЛИТЕТ|БАКАЛАВРИАТ|0.')][0].reset_index(drop = True)
second = df[~df[1].str.contains('Направленность', na = True)][1].reset_index(drop = True)
```

```python
z = []
x = []
l = []
k = []
h = []
def get_mos(first):
    
    for j in range(len(second)):
        if ((first[j].__contains__('БАКАЛАВРИАТ')) or (first[j].__contains__('СПЕЦИАЛИТЕТ'))):
                bak = first[j].capitalize()
                first = first.drop(j).reset_index(drop = True)
        if (second[j].count('\r') == 0):
            l.append(first[j].split('\r')[0])
            k.append(first[j].split('\r')[1])
            z.append(second[j].split('(')[0].strip())
            x.append(second[j].split('(')[-1].strip(')'))
            h.append(bak)

        else:
            if (len(re.findall(r'\([А-Я ]*?\)', second[j])) == 0):
                m = 1
            else: 
                m = len(re.findall(r'\([А-Я ]*?\)', second[j]))
                        
            for i in range(m):
                
                b = second[j].replace('\r',' ').split('(')[i]
                #b = second[j].split(re.findall(r'\([А-Я ]*?\)', second[j])[i])
                if len(re.findall('([А-Я][а-я]+)', second[j].replace('\r',' ').split('(')[i])) != 0:
                    if b.__contains__(')'):
                        #x.append(b.split(')')[0].strip())
                        z.append(second[j].split(re.findall(r'\([А-Я ]*?\)', second[j])[0])[i].replace('\r',' ').strip())
                        h.append(bak)
                        l.append(first[j].split('\r')[0])
                        x.append(re.findall(r'\([А-Я ]*?\)', second[j])[0].strip(')('))
                        #z.append(b.split(')')[1].strip())
                        k.append(first[j].split('\r',1)[1].replace('\r',' '))
                    else:
                        #x.append(second[j].split(re.findall(r'\([А-Я ]*?\)', second[j])[0])[i].replace('\r',' '))
                        x.append(second[j].replace('\r',' ').split('(')[-1].strip(')')) 
                        h.append(bak)
                        l.append(first[j].split('\r')[0])
                        z.append(b.strip())
                        k.append(first[j].split('\r',1)[1].replace('\r',' '))

get_mos(first)
mos = pd.DataFrame()
mos['код'] = l
mos['стандарт'] = k
mos['программа'] = z
mos['подразделение'] = x

#если нужно разделить на бакалавриат/специалитет
#mos['уровень'] = h 
mos = mos.drop_duplicates().reset_index(drop = True) #чтобы избавиться от повторных программ бюджет/договор или очка/заочка
mos
```

```python
dff.head(10)
fil = dff.iloc[:,0:1][~dff[0].str.contains("Код|Очная форма обучения|Очно-заочная форма обучения|Заочная форма обучения", na=True)].reset_index(drop = True)
fil = fil[ fil[0].str.contains('0.|ФИЛИАЛ|БАКАЛАВРИАТ|СПЕЦИАЛИТЕТ', na=False) ].reset_index(drop = True)#.iloc[:, 1:3]
st = fil[0]
fill = dff.iloc[:,0:3][~dff[0].str.contains("Код|БАКАЛАВРИАТ|Очная форма обучения|Очно-заочная форма обучения|СПЕЦИАЛИТЕТ|Заочная форма обучения", na=False)]
fill = fill[fill[2].str.contains("Образовательная|Совокупность", na=False)].reset_index()
```

```python
a = []
b = []
c = []
d = []
e = []

def get_fil(st):
    for i in range(0,len(fill.iloc[:,2])):
            if (st[i].__contains__('ФИЛИАЛ')):
                fili = st[i].capitalize()
                st = st.drop(i).reset_index(drop = True)
            if ((st[i].__contains__('БАКАЛАВРИАТ')) or (st[i].__contains__('СПЕЦИАЛИТЕТ'))):
                bak = st[i].capitalize()
                st = st.drop(i).reset_index(drop = True)
                
            if (fill.iloc[:,3][i].split('\r', 1)[0].__contains__('Образовательная') == True):
                if len(re.findall(r'([А-Я])', st[i].split('\r')[0])) == 0:
                    a.append(fill.iloc[:,1][i].split('\r')[0])
                    b.append(' '.join(st[i].split('\r')[1:(st[i]).count('\r')+1]).strip('*'))
                else: 
                    a.append(st[i].split(re.findall(r'([А-Я])', st[i])[0])[0])
                    b.append((re.findall(r'([А-Я])', st[i])[0] + st[i].split(re.findall(r'([А-Я])', st[i])[0])[1]).replace('\r', ' '))
                c.append((' '.join(' '.join(fill.iloc[:,3][i].split('\r')[1:fill.iloc[:,3][i].count('\r')+1]).split('\r',1))))
                d.append(fili)
                e.append(bak)
            elif (fill.iloc[:,3][i].split('\r', 1)[0].__contains__('Совокупность') == True):
                for j in range(1, fill.iloc[:,3][i].count('.')+1):
                    if len(re.findall(r'([А-Я])', st[i].split('\r')[0])) == 0:
                        a.append(fill.iloc[:,1][i].split('\r')[0])
                        b.append(' '.join(st[i].split('\r')[1:(st[i]).count('\r')+1]).strip('*'))
                    else: 
                        a.append(st[i].split(re.findall(r'([А-Я])', st[i])[0])[0])
                        b.append((re.findall(r'([А-Я])', st[i])[0] + st[i].split(re.findall(r'([А-Я])', st[i])[0])[1]).replace('\r', ' '))
                    c.append(fill.iloc[:,3][i].split('.')[j].replace('\r', ' ').strip('1234567890. '))
        
                    d.append(fili)
                    e.append(bak)
get_fil(st)
filiall = pd.DataFrame()
filiall['код'] = a
filiall['стандарт'] = b
filiall['программа'] = c
filiall['подразделение'] = d

#если нужно разделить на бакалавриат/специалитет
#filiall['уровень'] = e
filiall = filiall.drop_duplicates().reset_index(drop = True) #чтобы избавиться от повторных программ бюджет/договор или очка/заочка
filiall
```
